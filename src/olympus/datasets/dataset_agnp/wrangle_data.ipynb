```python
df = pd.read_csv('raw_data.csv')
print(df.shape)
df.head()
```

```python
feat_cols = df.columns[:6].tolist()
replace_cols = ['q_agno3', 'q_pva', 'q_tsc', 'q_seed', 'q_tot', 'spectrum_score']
```

```python
df = df.rename(columns={feat_col:replace_col for feat_col, replace_col in zip(feat_cols, replace_cols)})
print(df.shape)
df.head()
```

```python
wo_dup = df.iloc[:, :5].drop_duplicates(keep='last')
print(wo_dup.shape)
wo_dup.head()

    
```

```python
measurements = []

for ix, row in wo_dup.iterrows():
    row = row.to_dict()    
    sub = df[
        (df['q_agno3']==row['q_agno3'])&
        (df['q_pva']==row['q_pva'])&
        (df['q_tsc']==row['q_tsc'])&
        (df['q_seed']==row['q_seed'])&
        (df['q_tot']==row['q_tot'])
    ]
    avg_measurement = sub['spectrum_score'].mean()
    measurements.append(avg_measurement)
```

```python
wo_dup['spectrum_score'] = measurements
```

```python
print(wo_dup.shape)
wo_dup.head()
```

```python
wo_dup.to_csv('data.csv', index=False)
```
