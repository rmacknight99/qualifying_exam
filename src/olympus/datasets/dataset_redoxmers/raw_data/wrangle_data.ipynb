```python
# load config
import json

config = json.load(open('config.json', 'r'))
config
```

```python
# make the descriptors df
desc_dict = {'variable_name': [], 'option_name': [], 'descriptor_name': [], 'descriptor_value': []}


# group_type=variable_name,  lookup_label=option_name, desc=descriptor_name
def get_descriptors(group_type, lookup_label, desc_name):
    """ returns list of descriptor values for a given
    list of descriptor types
    """
    desc_vals = []
    val = df[df[f'{group_type}']==lookup_label][f'{group_type.split("_")[0]}_{desc_name}'].tolist()[0]
    return val



variable_names = [v['name'] for v in config['parameters']]
option_names   = [v['options'] for v in config['parameters']]
descriptor_names = [v['descriptors'] for v in config['parameters']]
```

```python
for variable_idx, variable_name in enumerate(variable_names):
    for option_name in option_names[variable_idx]:
        for descriptor_name in descriptor_names[variable_idx]:
            print(variable_name, option_name, descriptor_name)
            val = get_descriptors(variable_name, option_name, descriptor_name)
            
            desc_dict['variable_name'].append(variable_name)
            desc_dict['option_name'].append(option_name)
            desc_dict['descriptor_name'].append(descriptor_name)
            desc_dict['descriptor_value'].append(val)
```

```python
desc_df = pd.DataFrame(desc_dict)
print(desc_df.shape)
desc_df.head()
```
